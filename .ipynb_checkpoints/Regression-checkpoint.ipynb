```python
import numpy as np
import matplotlib.pyplot as plt
d = 1
N = 128
LR = 0.002
X = np.random.rand(N,d)
W = np.random.rand(d,1)
b = np.random.rand(N,1)


Y_pred = np.dot(X,W) + b

#創資料
Y = 1 + 2*np.dot(X, np.random.randint(1,8,d).reshape(d,1)) + 20*np.random.rand(N,1)
print(Y.shape)
#整個batch 的 MSE
def loss_func(Y_pred=[],Y=[]):
    return np.mean(np.square((Y_pred - Y)))

loss = []


for t in range(2000):
    #forward
    Y_pred = np.dot(X,W) + b
    
    #optimization step
    W = W - LR*np.dot(X.transpose(),(Y_pred-Y))
    b = b - LR*(Y_pred-Y)
    
    loss.append(loss_func(Y_pred,Y))
    
    if t%50 == 0:
        print('[t:{0}] loss: {1}'.format(t,loss[t]))

plt.plot(range(len(loss)),loss, label='Gradient Descent')
plt.legend(loc='upper right')
plt.xlabel('Steps')
plt.ylabel('Loss')
plt.ylim((0, 60))
```

```python
"""
Just use simple model for simple problem
"""
import numpy as np
import matplotlib.pyplot as plt

global N 
N = 1000
global d
d = 20
global LR
LR = 0.03
BATCH_SIZE = 20 #要被N整除,否則loader會出錯
EPOCH = 20

# dataset
x = np.random.rand(N,d)
y = 1 + 2*np.dot(x, np.random.randint(1,8,d).reshape(d,1)) + 20*np.random.rand(N,1)


def loader_func(x,y,batch_size,shuffle = True):
    split_num = N/batch_size
    x = np.split(x,split_num,0)
    y = np.split(y,split_num,0)
    loader = [(x_b,y_b) for x_b,y_b in zip(x,y)]
    if shuffle == True:
        random.shuffle(loader)
    return loader
    
loader = loader_func(x,y,BATCH_SIZE)
    
class regr(object):
    def __init__(self):
        self.weight = np.random.rand(d,1)
        self.bias = np.random.rand(1)
        self.grad = np.zeros((d+1,1))
        self.first_moment = np.zeros((d+1,1))
        self.velocity = np.zeros((d+1,1))
        self.squared_grad = np.zeros((d+1,1))
        self.time_step = 0

    def loss_func(self,x,y,W = None,b = None):
        if W is None:
            W = self.weight
        if b is None:
            b = self.bias
        y_hat = np.dot(x,W) + b
        y_hat = 1/(1+np.exp(-y_hat)) #sigmoid 
        return np.mean(np.square(y_hat - y))
    
    def numeric_grad(self,x,y):
        #zero grad
        self.grad = np.zeros((d+1,1))
        delta = 1e-4
        #grad = (Loss(wi = wi+delta)-Loss(wi = wi-delta))/(2*delta)
        for d_i in range(d):
            W_perturb_plus, W_perturb_minus = self.weight , self.weight
            W_perturb_plus[d_i] += delta
            W_perturb_minus[d_i] -= delta
            
            self.grad[d_i] = (self.loss_func(x, y, W = W_perturb_plus) - \
            self.loss_func(x, y, W = W_perturb_minus))/(2*delta)

            self.grad[-1] = (self.loss_func(x, y, b = self.bias + delta) - \
            self.loss_func(x, y, b = self.bias - delta))/(2*delta)
    
    def back_prop(self,x,y):
        '''backprop for linear regression'''
        self.grad = np.zeros((d+1,1))
        y_hat = np.dot(x,self.weight) + self.bias
        self.grad[:-1] = 2*np.dot(x.transpose(),y_hat - y)
        self.grad[-1] = 2*np.sum(y_hat - y)
    
    def optim(self,x,y,method = 'SGD',momentum = 0.8,alpha=0.9,betas=(0.9, 0.99)):
        self.numeric_grad(x,y)
        if method == 'SGD':
            self.weight = self.weight - LR*self.grad[:-1]
            self.bias = self.bias - LR*self.grad[-1]
        if method == 'Momentum':
            self.velocity = momentum*self.velocity - LR*self.grad
            #update weight
            self.weight = self.weight + self.velocity[:-1]
            self.bias = self.bias + self.velocity[-1]
        if method == 'AdaGrad':
            self.back_prop(x,y)
            #G hadamard product G
            self.squared_grad += np.square(self.grad)
            #calculate delta weight
            delta_weight = -LR*self.grad/(1e-7+np.sqrt(self.squared_grad))
            #apply update
            self.weight += delta_weight[:-1]
            self.bias += delta_weight[-1]
        if method == 'RMSprop':
            #accumulate squared_grad
            self.squared_grad = alpha*self.squared_grad + (1.-alpha)*np.square(self.grad)
            #calculate delta weight
            delta_weight = -LR*self.grad/(1e-8+np.sqrt(self.squared_grad))
            #apply update
            self.weight += delta_weight[:-1]
            self.bias += delta_weight[-1]
        if method == 'Adam':
            self.time_step += 1
            #second moment
            self.squared_grad = betas[1]*self.squared_grad + (1-betas[1])*np.square(self.grad)
            second_unbias = self.squared_grad/(1.-betas[1]**self.time_step)
            #first moment
            self.first_moment = betas[0]*self.grad + (1. - betas[0])*self.grad
            first_unbias = self.first_moment/(1-betas[0]**self.time_step)
            #calculate delta weight
            delta_weight = -LR*first_unbias/(1e-8 + np.sqrt(second_unbias))
            #apply update
            self.weight += delta_weight[:-1]
            self.bias += delta_weight[-1]
            

if __name__ == '__main__':
    # different regrs
    regr_SGD         = regr()
    regr_Momentum    = regr()
    regr_RMSprop     = regr()
    regr_Adam        = regr()
    regrs = [regr_SGD, regr_Momentum, regr_RMSprop, regr_Adam]

    # different optimization method
    optimizers = ["SGD", "Momentum", "RMSprop", "Adam"]

    losses_his = [np.zeros((EPOCH,)),np.zeros((EPOCH)),np.zeros((EPOCH)),np.zeros((EPOCH))] # record loss
    
    # training
    for epoch in range(EPOCH):
        if epoch % 5 == 0:
            print('Epoch: ', epoch)
        random.shuffle(loader)
        for step, (x_b, y_b) in enumerate(loader,1):          # for each training step
            for regr_i, optim_m, l_his in zip(regrs, optimizers ,losses_his):
                l_his[epoch] += regr_i.loss_func(x_b, y_b)/EPOCH #Average loss for each epoch
                regr_i.optim(x_b, y_b,method = optim_m) #update weight one step
            
                    
    for i, l_his in enumerate(losses_his):
        plt.plot(l_his, label=optimizers[i])
    plt.legend(loc='best')
    plt.xlabel('Steps')
    plt.ylabel('Loss')
    plt.show()
```

```python
"""
Appendix-PyTorch版
code參考吳莫凡的教學
"""
import torch
import torch.utils.data as Data
import torch.nn.functional as F
import matplotlib.pyplot as plt

# torch.manual_seed(1)    # reproducible

N = 1000
global d
d = 20
LR = 0.03
BATCH_SIZE = 30
EPOCH = 10

x = np.random.rand(N,d)
y = 1 + 2*np.dot(x, np.random.randint(1,8,d).reshape(d,1)) + 20*np.random.rand(N,1)
x = torch.tensor(x, dtype=torch.float32)
y = torch.tensor(y, dtype=torch.float32)

# put dateset into torch dataset
torch_dataset = Data.TensorDataset(x, y)
loader = Data.DataLoader(dataset=torch_dataset, batch_size=BATCH_SIZE, shuffle=True, num_workers=2,)


# default network
class Net(torch.nn.Module):
    def __init__(self, D_in = d, D_out = 1):
        super(Net, self).__init__()
        self.hidden = torch.nn.Linear(D_in, D_out)   # hidden layer
        
    def forward(self, x):
        x = self.hidden(x)     # activation function for hidden layer
        return x

if __name__ == '__main__':
    # different nets
    net_SGD         = Net()
    net_Momentum    = Net()
    net_RMSprop     = Net()
    net_Adam        = Net()
    nets = [net_SGD, net_Momentum, net_RMSprop, net_Adam]

    # different optimizers
    opt_SGD         = torch.optim.SGD(net_SGD.parameters(), lr=LR)
    opt_Momentum    = torch.optim.SGD(net_Momentum.parameters(), lr=LR, momentum=0.8)
    opt_RMSprop     = torch.optim.RMSprop(net_RMSprop.parameters(), lr=LR, alpha=0.9)
    opt_Adam        = torch.optim.Adam(net_Adam.parameters(), lr=LR, betas=(0.9, 0.99))
    optimizers = [opt_SGD, opt_Momentum, opt_RMSprop, opt_Adam]

    loss_func = torch.nn.MSELoss()
    losses_his = [[], [], [], []]   # record loss

    # training
    for epoch in range(EPOCH):
        print('Epoch: ', epoch)
        for step, (b_x, b_y) in enumerate(loader):          # for each training step
            for net, opt, l_his in zip(nets, optimizers, losses_his):
                output = net(b_x)              # get output for every net
                loss = loss_func(output, b_y)  # compute loss for every net
                opt.zero_grad()                # clear gradients for next train
                loss.backward()                # backpropagation, compute gradients
                opt.step()                     # apply gradients
                l_his.append(loss.data.numpy())     # loss recoder

    labels = ['SGD', 'Momentum', 'RMSprop', 'Adam']
    for i, l_his in enumerate(losses_his):
        plt.plot(l_his, label=labels[i])
    plt.legend(loc='best')
    plt.xlabel('Steps')
    plt.ylabel('Loss')
    plt.show()
```
